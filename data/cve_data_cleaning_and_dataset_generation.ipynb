```python
import pandas as pd
import numpy as np

year_file_name = '2017'

nvdcve = pd.read_json('./data/nvdcve-1.1-'+year_file_name+'.json')
nvdcve['CVE_Items'][0]

cve_df = pd.json_normalize(nvdcve['CVE_Items'])
cve_df.info()
```

```python
cve_df = cve_df.rename(columns={"cve.CVE_data_meta.ID": "CVE_ID", "cve.problemtype.problemtype_data": "problemtype", "cve.description.description_data":"description",
                  "impact.baseMetricV3.cvssV3.version": "cvssV3_version", 
                    "impact.baseMetricV3.cvssV3.vectorString": "cvssV3_vectorString",
                  "impact.baseMetricV3.cvssV3.attackVector": "cvssV3_attackVector",
                  "impact.baseMetricV3.cvssV3.attackComplexity": "cvssV3_attackComplexity",
                  "impact.baseMetricV3.cvssV3.privilegesRequired": "cvssV3_privilegesRequired",
                  "impact.baseMetricV3.cvssV3.userInteraction": "cvssV3_userInteraction",
                  "impact.baseMetricV3.cvssV3.scope": "cvssV3_scope",
                   "impact.baseMetricV3.cvssV3.confidentialityImpact": "cvssV3_confidentialityImpact",
                    "impact.baseMetricV3.cvssV3.integrityImpact": "cvssV3_integrityImpact",
                    "impact.baseMetricV3.cvssV3.availabilityImpact": "cvssV3_availabilityImpact",
                    "impact.baseMetricV3.cvssV3.baseScore": "cvssV3_baseScore",
                    "impact.baseMetricV3.cvssV3.baseSeverity": "cvssV3_baseSeverity",
                    "impact.baseMetricV3.exploitabilityScore": "V3_exploitabilityScore",
                    "impact.baseMetricV3.impactScore": "V3_impactScore"
                  })
cve_df.info()        
```

```python
cve_df = cve_df.iloc[:,[0,1,5,7,9]+list(np.arange(12,26))]
cve_df
```

```python
cve_df.info()
```

```python
for i in range(0, cve_df['description'].shape[0]):
    cve_df['description'][i] = cve_df['description'][i][0]
cve_description_data = pd.json_normalize(cve_df['description'])
cve_df['description'] = cve_description_data["value"]
```

```python
pd.set_option('display.max_rows', 100)

for i in range(0, cve_df['problemtype'].shape[0]):
    cve_df['problemtype'][i] = cve_df['problemtype'][i][0]

cve_problemtype_data = pd.json_normalize(cve_df['problemtype'])
cve_problemtype_data
```

```python
cve_problemtype_data['CWE1'] = "{'lang': 'None', 'value': 'None'}"
cve_problemtype_data['CWE2'] = "{'lang': 'None', 'value': 'None'}"
cve_problemtype_data['nb_CWE'] = 0

for i in range(0, cve_problemtype_data.shape[0]):
    nb_element = len(cve_problemtype_data['description'][i])
    cve_problemtype_data['nb_CWE'][i] = nb_element
    if nb_element == 1:
        cve_problemtype_data['CWE1'][i] = cve_problemtype_data['description'][i][0]
    if nb_element > 1:
        cve_problemtype_data['CWE1'][i] = cve_problemtype_data['description'][i][0]
        cve_problemtype_data['CWE2'][i] = cve_problemtype_data['description'][i][1]
    
    
#cve_2020_problemtype_data = pd.json_normalize(cve_2020['cve.problemtype.problemtype_data'])

cve_problemtype_data
```

```python
import ast

CWE1_dict_list = list(cve_problemtype_data.CWE1)

CWE1 = []

for i, element in enumerate(CWE1_dict_list):
    if type(element) is str:
        element = ast.literal_eval(element)
        #print(i, dict(element)['value'])
        CWE1.append(dict(element)['value'])
    else:
        #print(i, dict(element)['value'])
        CWE1.append(dict(element)['value'])
CWE1
```

```python
import ast

CWE2_dict_list = list(cve_problemtype_data.CWE2)

CWE2 = []

for i, element in enumerate(CWE2_dict_list):
    if type(element) is str:
        element = ast.literal_eval(element)
        #print(i, dict(element)['value'])
        CWE2.append(dict(element)['value'])
    else:
        #print(i, dict(element)['value'])
        CWE2.append(dict(element)['value'])
CWE2
```

```python
cve_df['nb_CWE'] = cve_problemtype_data['nb_CWE']
cve_df['CWE1'] = CWE1
cve_df['CWE2'] = CWE2

cve_df.drop(columns=['problemtype'], inplace=True)
cve_df
```
